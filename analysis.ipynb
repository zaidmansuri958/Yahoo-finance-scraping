```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df=pd.read_excel("stock_data.xlsx")
```

```python
df.head()
```

```python
df = df.drop_duplicates(subset=["name"])
```

```python
top5_highest_marketcap = df.sort_values("market_cap_B", ascending=False).head(5)
```

```python
top5_highest_marketcap
```

```python
plt.figure(figsize=(20,10))
plt.bar(top5_highest_marketcap["name"],top5_highest_marketcap["market_cap_B"],color="green")
plt.xlabel("Company name",size=24)
plt.ylabel("Market Cap(Billion)",size=24)
plt.title("Top 5 Companies By Market Cap (Billion)",size=28,color="blue")
plt.xticks(size=16)
plt.yticks(size=16)
plt.show()
```

```python
bottom5_marketcap=df.sort_values("market_cap_B",ascending=True).head(5)
```

```python
bottom5_marketcap
```

```python
plt.figure(figsize=(20,10))
plt.bar(bottom5_marketcap["name"],bottom5_marketcap["market_cap_B"],color="red")
plt.xlabel("Company Name",size=24)
plt.ylabel("Market cap(Billion)",size=24)
plt.title("Bottom 5 Companies By Market Cap (Billion)",size=28,color="blue")
plt.xticks(size=16)
plt.yticks(size=16)
plt.show()
```

```python
top_5_highest_price=df.sort_values("price_usd",ascending=False).head(5)
```

```python
plt.figure(figsize=(10,6))
plt.plot(top_5_highest_price["name"],top_5_highest_price["price_usd"])
plt.grid(True)
plt.xlabel("Company Name")
plt.ylabel("Stock Price(USD)")
plt.title("Top 5 Companies by share price",size=28,color="blue")
plt.xticks(rotation=90)
plt.plot()
```

```python
top5_lowest_price=df.sort_values("price_usd",ascending=True).head(5)
```

```python
top5_lowest_price
```

```python
plt.figure(figsize=(10,6))
plt.plot(top5_lowest_price["name"],top5_lowest_price["price_usd"])
plt.xlabel("Company name")
plt.ylabel("Price USD")
plt.title("Bottom 5 Companies by share price",size=28,color="blue")
plt.grid(True)
plt.show()
```

```python
top5_highest_pe=df.sort_values("pe_ratio",ascending=False).head(5)
```

```python
top5_highest_pe
```

```python
plt.figure(figsize=(10,6))
plt.barh(top5_highest_pe["name"],top5_highest_pe["pe_ratio"])
plt.xlabel("PE ratio")
plt.title("Top 5 Companies by PE ratio",size=28,color="blue")
plt.ylabel("Company name")
plt.show()
```

```python
top5_lowest_pe=df.sort_values("pe_ratio",ascending=True).head(5)
```

```python
top5_lowest_pe
```

```python
plt.figure(figsize=(10,6))
plt.barh(top5_lowest_pe["name"],top5_lowest_pe["pe_ratio"])
plt.xlabel("PE ratio")
plt.ylabel("Company name")
plt.title("Bottom 5 Companies by PE ratio",size=28,color="blue")
plt.show()
```
